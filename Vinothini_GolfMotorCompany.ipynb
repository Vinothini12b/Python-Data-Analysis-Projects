***`Task 1: Import Pandas Library and create a dataframe named “tyresales” from a given URL and then print it. URL - https://raw.githubusercontent.com/Invact-Abhay/DO9/main/tyresale.csv`***

```python
import pandas as pd
tyresales=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DO9/main/tyresale.csv")
tyresales
```

***`Task 2: Clean the customer name by using strip and title function and removing '@' and any space between the name and give name to this column - ‘CustomerNameClean’ then retrieve ‘CustomerNameClean’ in Alphabetical order. ‘CustomerNameClean’ should look like this - AlphaMotorsPvtLtd.`***

```python
tyresales["CustomerNameClean"]=tyresales["CustomerName"].str.strip()\
                                                        .str.replace('@','')\
                                                        .str.title()\
                                                        .str.replace(' ','')+'.'
tyresales.sort_values("CustomerNameClean",ascending=True)
```

***`Task 3: Calculate the Sales for every customer by calculating Sales column`***

```python
tyresales["Sales"]=tyresales["UnitsSold"]*tyresales["PerUnitPrice"]
tyresales
```

**`Task 4: Find the total Sales.`**

```python
TotalSales=tyresales["Sales"].sum()
TotalSales
```

***`Task 5: Take the user Input for CustomerNameClean and retrieve the data for 'Alpha Motors Pvt Ltd'.`***

```python
customer_input=input("Enter the Customer Name:")
customer_data = tyresales[tyresales['CustomerNameClean'] == customer_input]
if not customer_data.empty:
    print("Data for", customer_input)
    print(customer_data)
else:
    print("No data found for", customer_input)
```

***`Task 6: Find the average SalesCommission% by calculating the SalesCommission% column.SalesCommission% = SalesCommissionPaid / Sales * 100`***

```python
tyresales["SalesCommission%"]=(tyresales["SalesCommissionPaid"] / tyresales["Sales"] )* 100
average_sales_commission = tyresales["SalesCommission%"].mean()
average_sales_commission
```

**`Task 7: Take the user Input for Zone and Find the average SalesCommission% of Central Zone.`**

```python
zone_input = input("Enter the Zone: ")
average_sales_commission_zone = tyresales[tyresales['Zone'] == zone_input]['SalesCommission%'].mean()
if average_sales_commission_zone:
    print(f"Average Sales Commission% for {zone_input} Zone:", average_sales_commission_zone)
else:
    print(f"No data found for {zone_input} Zone.")
```
