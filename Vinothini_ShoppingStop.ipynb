**Task 13:**

Store the user input for the month in the variable named as "Input_Month1". Text to be shown to the user "Enter the Month Name : "

Store the user input for the ProductCategory in the variable named as "Input_ProductCategory". Text to be shown to the user "Enter the Category Name : "

Filter the sales dataframe for Month to be taken from user input and Product Category to be taken from user input. Store the resultant data in "Sales3" dataframe.

Sort the "Sales3" by arranging

"TotalRevenue column in descending order

At the end the user input should be-

Month (Input_Month1) - Enter the name of Month: July

ProductCategory (Input_ProductCategory) - Enter the name of ProductCategory: Electronics

```python
Input_Month1=input("Enter the Month Name: ")
Input_ProductCategory=input("Enter the Category Name : ")
Sales3=salesdata.query('Month==@Input_Month1 and ProductCategory==@Input_ProductCategory')
Sales3=Sales3.sort_values("TotalRevenue",ascending=False)
Sales3
```

**Task 14:**

Store the user input for the region in the variable named as "Input_Region3". Text to be shown to the user "Enter the Region Name : "

Store the user input for the ProductCategory in the variable named as "Input_ProductCategory1". Text to be shown to the user "Enter the ProductCategory Name : " Filter the sales dataframe for Region to be taken from user input and ProductCategory to be taken from user input. Store the resultant data in "Sales4" dataframe.

Create a pivot table using the Sales4 dataframe named pivot_2 to show the ProductName-wise total revenue and total unit sold for each of the ProductCategory. Use index as ProductName, columns as ProductCategory and values as TotalRevenue and UnitsSold. Print this pivot table. This will show the data for the ProductCategory and Region selected by the user.

At the end the user input should be-

Region (Input_Region3) - Enter the region name : Asia

ProductCategory (Input_ProductCategory1) - Enter the ProductCategory name : Clothing

```python
Input_Region3=input("Enter the Region Name : ")
Input_ProductCategory1=input("Enter the ProductCategory Name : " )
Sales4=salesdata.query('Region==@Input_Region3 and ProductCategory==@Input_ProductCategory1')
Pivot_2= pd.pivot_table(Sales4,index="ProductName",columns="Region",values=["TotalRevenue", "UnitsSold"],aggfunc={"UnitsSold":["sum"],"TotalRevenue":["sum"]})
Pivot_2
```
