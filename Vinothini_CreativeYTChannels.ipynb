**Task 1**


Import Pandas  and matplotlib Library.
Create a dataframe named "ytdata" from a given URL and then print it.

URL - https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/FE%20Final%20Final.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
ytdata=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/FE%20Final%20Final.csv")
ytdata
```

**Task 2**

Convert impression, videoviews and subscribers into millions (in '000000 format) and round to 2 decimal places  and name these columns  -


*   impression(in '000000)
*   videoviews(in '000000)
*   subscribers(in '000000)

Display the 5 rows of DataFrame

```python
ytdata["impression(in '000000)"]=(ytdata["impression"]/1000000).round(2)
ytdata["videoviews(in '000000)"]=(ytdata["videoviews"]/1000000).round(2)
ytdata["subscribers(in '000000)"]=(ytdata["subscribers"]/1000000).round(2)
ytdata.head()
```

**Task 3:**

Extract the year and month from "started" (Channel Start Date) in separate column named "year" and "month" respectively.

Display the 5 rows of DataFrame

```python
ytdata["started"]=pd.to_datetime(ytdata["started"])
ytdata["Year"]=ytdata["started"].dt.year
ytdata["Month"]=ytdata["started"].dt.month
ytdata.head()
```

**Task 4:**

Drop the following columns -
'impression', 'videoviews','subscribers', 'started'

Display the 5 rows of DataFrame

```python
ytdata.drop(columns=['impression', 'videoviews','subscribers', 'started'],inplace=True)
ytdata.head()
```

# Bar Chart

**Task 5:**


Create a pivot table from the ytdata DataFrame that counts the number of channels ("youtuber") created each year. Store this pivot table in Pivot_bar and reset its index. Then, print the Pivot_bar.

```python
Pivot_bar= pd.pivot_table(ytdata,values=["youtuber"],index=["Year"],aggfunc={"youtuber":"count"})
Pivot_bar.reset_index(inplace=True)
Pivot_bar
```


**Task 6:**

Create a bar plot using the Pivot_bar DataFrame, with years on the x-axis and the count of channels ("youtuber") on the y-axis. Then, display the plot.

```python
plt.bar(x=Pivot_bar["Year"],height=Pivot_bar["youtuber"])
plt.show()
```

**Task 7:**

Create a bar plot using the Pivot_bar DataFrame, with years on the x-axis and the "youtuber" on the y-axis.

Label the x-axis as "Year"

and the y-axis as "Count of channels".

Add the title "Year wise channels created",

and rotate the x-axis labels by 45 degrees.

Then, display the plot.

```python
plt.bar(x=Pivot_bar["Year"],height=Pivot_bar["youtuber"])
plt.xlabel("Year")
plt.ylabel("Count of channels")
plt.title("Year wise channels created")
plt.xticks(Pivot_bar["Year"],rotation=45)
plt.show()
```

# Pie Chart

**Task 8:**

Create a pivot table from the ytdata DataFrame that counts the number of channels (youtuber) for each region.

Store this pivot table in Pivot_Pie and reset its index.

Display the resulting DataFrame.

```python
Pivot_Pie=pd.pivot_table(ytdata,index=["region"],values=["youtuber"],aggfunc={('youtuber'):"count"}).reset_index()
Pivot_Pie
```

**Task 9:**

Create a pie chart using the Pivot_Pie DataFrame, with the count of channels (youtuber) for each region.

label the slices with region,

and set the title to "Share of region".

Then, display the chart.

```python
plt.pie(Pivot_Pie["youtuber"],labels= Pivot_Pie["region"])
plt.title("Share of region")
plt.show()
```

**Task 10:**

Create a pie chart using the Pivot_Pie DataFrame, with the count of channels (youtuber) for each region.

label the slices with region,

and display the percentage of each slice with no decimal places.

Set the title to "Share of region"

and then display the chart.

```python
plt.pie(Pivot_Pie["youtuber"],labels= Pivot_Pie["region"],autopct="%0.0f%%")
plt.title("Share of region")
plt.show()
```

# Scatter Plot

**Task 16:**

Create a scatter plot to visualize the relationship between videoviews and impression of channels with region as "Asia" in the ytdata DataFrame. Then, display the plot.

```python
videoviews_Asia= ytdata.query('region=="Asia"')["videoviews(in '000000)"]
Impressionviews_Asia= ytdata.query('region=="Asia"')["impression(in '000000)"]
plt.scatter(videoviews_Asia,Impressionviews_Asia)
plt.show()
```

**Task 17:**

Create a scatter plot to visualize the relationship between videoviews and impression of channels with region as "Asia" and "Europe" in the ytdata DataFrame. Then, display the plot.








```python
videoviews_Asia= ytdata.query('region=="Asia"')["videoviews(in '000000)"]
Impressionviews_Asia= ytdata.query('region=="Asia"')["impression(in '000000)"]
plt.scatter(videoviews_Asia,Impressionviews_Asia)
videoviews_Europe= ytdata.query('region=="Europe"')["videoviews(in '000000)"]
Impressionviews_Europe= ytdata.query('region=="Europe"')["impression(in '000000)"]
plt.scatter(videoviews_Europe,Impressionviews_Europe)
plt.show()
```

**Task 18:**

Add the label and titles in the scatter plot created above as per the details provided below:

For Asia scatter plot -   legend - 'Asia'

For Europe scatter plot -   legend - 'Europe'

Label the x-axis as "View Count(in millions)

and the y-axis as "Impression Count(in millions)"

and the title "Asia vs Europe".

and place the legend on upper left and legend title is 'Region'

```python
videoviews_Asia= ytdata.query('region=="Asia"')["videoviews(in '000000)"]
Impressionviews_Asia= ytdata.query('region=="Asia"')["impression(in '000000)"]
plt.scatter(videoviews_Asia,Impressionviews_Asia,label="Asia")
videoviews_Europe= ytdata.query('region=="Europe"')["videoviews(in '000000)"]
Impressionviews_Europe= ytdata.query('region=="Europe"')["impression(in '000000)"]
plt.scatter(videoviews_Europe,Impressionviews_Europe,label="Europe")
plt.xlabel("View Count(in millions)")
plt.ylabel("Impression Count(in millions)")
plt.title("Asia vs Europe")
plt.legend(title="Region",loc="upper left")
plt.show()
```
